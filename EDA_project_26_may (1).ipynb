```python
df[['Minimum hours', 'Maximum hours']] = df["How many hours you want to spend on learning new technologies every week?"].str.split
df['Minimum hours'] = df['Minimum hours'].str.replace(' hours', '')
df['Maximum hours'] = df['Maximum hours'].str.replace(' hours', '')
df['Minimum hours'] = pd.to_numeric(df['Minimum hours'])
df['Maximum hours'] = pd.to_numeric(df['Maximum hours'])
df
```

```python
# task_3
```

```python
df["How many years are you freelancing?"]=df["How many years are you freelancing?"].replace("21+ years","21-22 years")
df[['Minimum years', 'Maximum years']] = df["How many years are you freelancing?"].str.split('-', expand=True)
df['Minimum years'] = df['Minimum years'].str.replace(' years', '')
df['Maximum years'] = df['Maximum years'].str.replace(' years', '')
df['Minimum years'] = pd.to_numeric(df['Minimum years'],errors='coerce')
df['Maximum years'] = pd.to_numeric(df['Maximum years'],errors='coerce')
df
```

```python
# task 4
```

```python
df = df.replace({'\+\+': '_Plus_Plus', '#': '_Sharp'},regex=True)
df
```

```python
data = df[['Age group you belong to?', 'What the average hours your work per week?']]

plt.figure(figsize=(10, 8))
sns.barplot(data=data, x='Age group you belong to?', y='What the average hours your work per week?',ci=None)
plt.title('Average Weekly Work Hours by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Average Work Hours per Week')
plt.xticks(rotation=45)
plt.show()
```

```python
top_programming = df['What Programming your most comfortable?'].value_counts()[:5]
top_programming.plot(kind='bar',color='green')
plt.title('Most Known Top 5 Programming Language');
```

```python
Learning_counts = df['How Did you Learn IT and Computer?'].value_counts()
Learning_counts
plt.pie(Learning_counts, labels=Learning_counts, autopct='%.1f%%')
plt.title('Learnt IT and Computer From ');
```

```python
data = df[['How satisfied are you with your freelancing job?', 'Which of these front-end frameworks you are good at?']]

# Filter the top 5 front-end frameworks
top_frameworks = data['Which of these front-end frameworks you are good at?'].value_counts().head(5).index
filtered_data = data[data['Which of these front-end frameworks you are good at?'].isin(top_frameworks)]

# Create a cluster bar graph
plt.figure(figsize=(10, 6))
sns.countplot(data=filtered_data, x='How satisfied are you with your freelancing job?',hue="Which of these front-end frameworks you are good at?")
plt.title('Cluster Bar Graph: Satisfaction and Front-end Frameworks')
plt.xlabel('Satisfaction')
plt.ylabel('Count')
plt.xticks(rotation=45)  
plt.legend(title='Front-end Frameworks');
```

```python
grouped_data=df.groupby("Age group you belong to?")["What the average hours your work per week?"].mean()
```

```python
sns.barplot(grouped_data.index,grouped_data,color="black")
plt.xticks(rotation=90)
plt.show()
```
